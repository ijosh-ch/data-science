### Simulating Coin Toss 

Tossing a coin multiple times will generate a binomial distribution. If the probability of heads is p, the coin is tossed n times, the number of heads for each n count is given by the binomial distribution.


```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
np.random.seed(30)
```

Tossing a fair coin once

```python
n=1
p=0.5 # fair coin
np.random.binomial(n,p)
```

If the result is 1 then it is a heads, or else tails

If we toss it 10 times: 

```python
n=10
p=0.5 # fair coin
np.random.binomial(n,p)
```

It gives number of times heads show up

We tossed a single coin 10 times. This is a single experiment. We can repeat the experiment multiple times. Let us repeat the experiment 10 times.

```python
n=10
p=0.5
# let us repeat our experiment for 10 times
size=10
x=np.random.binomial(n=n, p=p, size=size)
x
```

In the multiple experiments, we can count how many heads in each of the individual experiment. 


```python
[np.equal(x,i).sum() for i in range(n)]
```

If the binomial experiment generated:  array([4, 6, 4, 2, 2, 6, 6, 7, 4, 2])

And the we counted the number of heads: [0, 0, 3, 0, 3, 0, 3, 1, 0, 0].

The results should be understood as:

The first number 0: in 0 of the experiment we have 0 heads. Second 0: in 0 of the experiments we have 1 head. Third 3: in 3 of the experiment we have 2 heads. Fourth 0: in 0 experiments we have 3 heads. Fifth 3: in 3 experiments we have 4 heads. And so on.

Instead of counts, we can calculate the probability instead (by dividing by the total)

```python
[np.equal(x,i).mean() for i in range(n)]
```

Let's consider a larger experiment run, say 100 times.


```python
n=10
p=0.5
# let us repeat our experiment for 10 times
size=100
x=np.random.binomial(n=n, p=p, size=size)
x
```

```python
prob=[np.equal(x,i).mean() for i in range(n)]
prob
```

You can see that 4-6 heads are the most likely. Lets do a plot.

```python
fig, ax = plt.subplots(1, 1)
ax.xaxis.set_ticks(range(n))
ax.set_xlabel('Number of Heads',fontsize=14)
ax.set_ylabel('Probability',fontsize=14)

ax.plot(list(range(n)), prob, color='blue', marker='o')

plt.show()
```

We overlay the binomial distribution over the coin toss experiment.
